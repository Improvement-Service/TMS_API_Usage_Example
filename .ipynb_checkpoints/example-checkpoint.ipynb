# Tell Me Scotland API

## config

In the config.ini set the missing fields with your details 

[service] <br> 
client_id= < client_id > <br> 
client_secret= < client_secret > <br> 
scope=['tellme'] <br> 
token_url=https://www.tellmescotland.gov.uk/oauth2/token <br> 
api_url=https://www.tellmescotland.gov.uk/api/v1/ <br> 
username= < username > <br> 
password= < password > <br> 
notices=https://www.tellmescotland.gov.uk/api/v1/notices.json <br> 
organisations=https://www.tellmescotland.gov.uk/api/v1/organisations.json <br> 
notice_types=https://www.tellmescotland.gov.uk/api/v1/notice_types.json <br> 


## setup

```python
#import required modules
from tms.service import *
from tms.base import Configuration
import json
```

```python
c = Configuration()
client = SetUp()
token = client.get_token()
header = client.create_header()
rest = RestRequests(token, header)
```

## Examples of getting data from api

```python
organisations_url = c.get_configuration_for('service', 'organisations')
limit = 1
url = f'{organisations_url}?limit={limit}'
print(url)
data = rest.get(url)
print(json.dumps(data, indent=2))
```

```python
notices_url = c.get_configuration_for('service', 'notices')
limit = 5
url = f'{notices_url}?limit={limit}'
print(url)
data = rest.get(url)
print(json.dumps(data, indent=2))
```

```python
notice_types_url = c.get_configuration_for('service', 'notice_types')
limit = 5
print(notice_types_url)
data = rest.get(notice_types_url)
print(json.dumps(data, indent=2))
```

### adding filters to requests



```python
notices_url = c.get_configuration_for('service', 'notices')
limit = 5
orid = 12
url = f'{notices_url}?orid={orid}&limit={limit}'
print(url)
data = rest.get(url)
print(json.dumps(data, indent=2))
```

```python
notices_url = c.get_configuration_for('service', 'notices')
limit = 5
ntid = 13
url = f'{notices_url}?ntid={ntid}&limit={limit}'
print(url)
data = rest.get(url)
print(json.dumps(data, indent=2))
```
